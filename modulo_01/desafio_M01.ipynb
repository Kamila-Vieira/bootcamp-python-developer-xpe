6. Considere a seguinte agenda de disponibilidade de atendimento de uma cl√≠nica com v√°rios m√©dicos, cada um de uma especialidade distinta. Esta cl√≠nica possu√≠ apenas tr√™s consult√≥rios, e, por isso, no m√°ximo tr√™s m√©dicos podem atender por dia:

image.png

A cl√≠nica percebeu que os pacientes que desejam se consultar com mais de um m√©dico sempre solicitam que o atendimento seja realizado no mesmo dia. Por exemplo, um paciente gostaria de se consultar, no mesmo dia, com o ortopedista e com o neurologista, ent√£o os dias dispon√≠veis ser√£o: ter√ßa e quinta. Outro exemplo √© um paciente que gostaria de se consultar com o dermatologista, o
neurologista e o psiquiatra. O √∫nico dia dispon√≠vel ser√° a sexta.


Entretanto, essas solicita√ß√µes est√£o causando dor de cabe√ßa, pois √© necess√°rio verificar manualmente a disponibilidade. Voc√™, como um(a) aspirante a desenvolvedor Python, aceitou o desafio de ajudar a automatizar essa tarefa. Para tanto, ir√° desenvolver duas fun√ß√µes que receber√£o a rela√ß√£o de disponibilidade de cada m√©dico e que retornar√£o os dias dispon√≠veis:


```python
# rela√ß√£o de dias da semana que cada m√©dico atende
cardiologista = {'terca', 'quarta'}
ortopedista = {'terca', 'quinta'}
dermatologista = {'segunda', 'quarta', 'sexta'}
neurologista = {'terca', 'quinta', 'sexta'}
psiquiatra = {'segunda', 'quarta', 'sexta'}

# Calcula quais os dias poss√≠veis para dois m√©dicos
def disp_dois_especialistas(medico01, medico02):
    I_med1_med2 = medico01 & medico02
    return I_med1_med2
    
# Calcula quais os dias poss√≠veis para tr√™s m√©dicos
def disp_tres_especialistas(medico01, medico02, medico03):
    I_med1_med2_med3 = medico01 & medico02 & medico03
    return I_med1_med2_med3
```

Assim, para os exemplos anteriores, as fun√ß√µes dever√£o ter os seguintes retornos:

```python
# M√©dicos solicitados: ortopedista e neurologista
print(disp_dois_especialistas(ortopedista, neurologista))
# RETORNA: ter√ßa e quinta

# M√©dicos solicitados: dermatologista, neurologista, psquiatra
print(disp_tres_especialistas(dermatologista, neurologista, psiquiatra))
# RETORNA: sexta
```
