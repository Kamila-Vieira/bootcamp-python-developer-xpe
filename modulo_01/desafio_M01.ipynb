image.png

A clÃ­nica percebeu que os pacientes que desejam se consultar com mais de um mÃ©dico sempre solicitam que o atendimento seja realizado no mesmo dia. Por exemplo, um paciente gostaria de se consultar, no mesmo dia, com o ortopedista e com o neurologista, entÃ£o os dias disponÃ­veis serÃ£o: terÃ§a e quinta. Outro exemplo Ã© um paciente que gostaria de se consultar com o dermatologista, o
neurologista e o psiquiatra. O Ãºnico dia disponÃ­vel serÃ¡ a sexta.


Entretanto, essas solicitaÃ§Ãµes estÃ£o causando dor de cabeÃ§a, pois Ã© necessÃ¡rio verificar manualmente a disponibilidade. VocÃª, como um(a) aspirante a desenvolvedor Python, aceitou o desafio de ajudar a automatizar essa tarefa. Para tanto, irÃ¡ desenvolver duas funÃ§Ãµes que receberÃ£o a relaÃ§Ã£o de disponibilidade de cada mÃ©dico e que retornarÃ£o os dias disponÃ­veis:


```python
# relaÃ§Ã£o de dias da semana que cada mÃ©dico atende
cardiologista = {'terca', 'quarta'}
ortopedista = {'terca', 'quinta'}
dermatologista = {'segunda', 'quarta', 'sexta'}
neurologista = {'terca', 'quinta', 'sexta'}
psiquiatra = {'segunda', 'quarta', 'sexta'}

# Calcula quais os dias possÃ­veis para dois mÃ©dicos
def disp_dois_especialistas(medico01, medico02):
    I_med1_med2 = medico01 & medico02
    return I_med1_med2
    
# Calcula quais os dias possÃ­veis para trÃªs mÃ©dicos
def disp_tres_especialistas(medico01, medico02, medico03):
    I_med1_med2_med3 = medico01 & medico02 & medico03
    return I_med1_med2_med3
```

Assim, para os exemplos anteriores, as funÃ§Ãµes deverÃ£o ter os seguintes retornos:

```python
# MÃ©dicos solicitados: ortopedista e neurologista
print(disp_dois_especialistas(ortopedista, neurologista))
# RETORNA: terÃ§a e quinta

# MÃ©dicos solicitados: dermatologista, neurologista, psquiatra
print(disp_tres_especialistas(dermatologista, neurologista, psiquiatra))
# RETORNA: sexta
```
