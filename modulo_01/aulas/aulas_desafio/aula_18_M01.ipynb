Argumentos das Funções

```python
# calcula a área de um triângulo: (b x h) / 2 
def area_triangulo(base, altura): 
    return (base * altura) / 2

print(area_triangulo(5, 10))
```

É possível também passar os valores dos argumentos em uma ordem diferente, mas neste caso precisamos declarar o nome dos argumentos na chamada da função, conforme o exemplo a seguir.

```python
print(area_triangulo(altura=10, base=5))
```

Se passarmos um número diferente de argumentos, o interpretador do Python irá retornar um erro.

```python
print(area_triangulo(5))
```

Podemos atribuir um valor padrão para um argumento por meio do operador de atribuição (=)

```python
def exibe_pessoa(nome, idade=30): 
    print(f'Meu nome é {nome} e tenho {idade} anos.')
```

```python
exibe_pessoa('Antônio') 
exibe_pessoa('Antônio', 36)
```
