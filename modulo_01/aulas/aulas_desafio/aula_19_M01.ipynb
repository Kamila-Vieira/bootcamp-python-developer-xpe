Importação do múdulo

```python
import areas

areas.triangulo(5, 8)
```

Se desejássemos importar todas as declarações, seria necessário a importação do código abaixo. Desta forma, teremos acesso direto a todos as declarações do módulo, inclusive a variável PI.

```python
from areas import * 

print(triangulo(5, 8))
print(quadrado(6))
PI
```

Importação de apenas algumas funções, neste caso, somente quadrado() e circulo() estarão disponíveis para utilização no código.

```python
from areas import quadrado, circulo

print(quadrado(8))
print(circulo(8, 2))
```

Entretanto, existem situações em que a importação de declarações diretamente não é recomendada, pois, ao fazer isso, perdemos a rastreabilidade de qual módulo está provendo uma função.

Também é possível atribuirmos um novo nome para o módulo quando realizamos a importação, da seguinte forma:

```python
import areas as ar

print(ar.triangulo(5, 8))
print(ar.quadrado(6))
```
