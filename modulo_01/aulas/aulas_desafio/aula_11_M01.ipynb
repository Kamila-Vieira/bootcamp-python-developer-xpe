# Estruturas de Dados em Python


Tipos de estrutura de dados

```python
# Lista => tipo list => Ordenada e mutável.
list_example = [1, 2, 3, 4]
print(list_example)
# Tupla => tipo tuple => Ordenada e imutável.
tuple_example = (1, 2, 3, 4)
print(tuple_example)
# Conjunto => tipo set => Não-ordenada, mutável e valores únicos.
set_example = {1, 2, 3, 4}
print(set_example)
# Dicionário => tipo dict => Mapeamento (key, value), não-ordenado e mutável.
dict_example = {'a':1, 'b':2, 'c':3}
print(dict_example)
```
