### Estruturas de Repetição

While

```
while <expressão>:
    <bloco de código a ser repetido enquanto a expressão for verdadeira>
```

````python
# exemplo com n = 15 
n = 15 
soma = 0
contador = 0 

while contador <= n: 
    soma = soma + contador 
    contador = contador + 1 

print(f'A soma dos primeiros {n} inteiros é {soma}')

# exemplo sem o contador
soma = 0

while n >= 0: 
    soma = soma + n 
    n = n - 1 

print(f'A soma dos primeiros inteiros é {soma}')
````

for … in

```
for <item> in <sequência>:
    <bloco de código a ser repetido para cada item da sequência>
```

````python
# for … in
# for <item> in <sequência>: 
#     <bloco de código a ser repetido para cada item da sequência>

# exemplo de palavra: araraquara 
p = 'araraquara' 
contador = 0 
for letra in p: 
    if letra == 'a': 
        contador = contador + 1 
        
print(f"A palavara {p} possuí {contador} letras 'a'")
````

range

```range(inicio, fim, incremento)```

````python
range(5) # 0, 1, 2, 3, 4 
range(3, 8) # 3, 4, 5, 6, 7 
range(0, 21, 5) # 0, 5, 10, 15, 20
range(2, 10, 2) # 2, 4, 6, 8 

# exemplo com n = 15 
n = 15 
soma = 0 
for num in range(n + 1): 
    soma = soma + num 

print(f'A soma dos primeiros {n} inteiros é {soma}')
````

**Foi necessário utilizar a função range com n+1 (range(n + 1). Se fosse utilizado apenas n como argumento (range(n)), o resultado seria a soma de apenas os 14 primeiros inteiros.**

Interrupção da Estruturas de Repetição (for ... in e while)

```
for <item> in <sequência>: 
    <bloco de código a ser repetido> 
    if <expressao>:
        <outro bloco de código> 
        break
```

E

```
while <condição de repetição>: 
    <bloco de código a ser repetido>
    if <condição de parada>:
        <outro bloco de código> 
        break
```

````python
# Realiza a busca na faixa de 250 e 10000 
for num in range(250, 10000): 
    # verifica se o número é divisível por 21 
    if num % 21 == 0: 
        print('Número encontrado:', num) 
    # se for divísivel por 21, interrompe a busca 
        break
````

````python
# Realiza a busca na faixa de 250 e 2000 
num = 250
while num < 2000: 
    # verifica se o número é divisível por 21 
    if num % 21 == 0: 
        print('Número encontrado:', num) 
    # se for divísivel por 21, interrompe a busca 
        break
    num += 1
````
