Booleanos

Uma vez que o tipo bool assume os valores True ou False, matematicamente são representados por 1 e 0 (respectivamente)

```python
print(10 + False) # Equivale a 10 + 0 
print(True + 5) # Equivale a 1 + 5 
print(True + True + False) # Equivale a 1 + 1 + 0
```

Regras de precedência

PEMDAS: Parênteses > Exponenciação > Multiplicação > Divisão > Adição > Subtração

image.png

```python
B = 15 # Base maior (B)
b = 10 # Base menor (b)
h = 8 # Altura (h)

# Calcula a área do Trapézio
area = B + b * h / 2
print('Área do Trapézio: ', area)
```

Observe que o resultado encontrado (55.0) não corresponde ao valor esperado para 𝐵 = 15, 𝑏 = 10 e ℎ = 8. Conforme o exemplo da figura, o valor correto seria 100.0!

O erro foi causado pela ordem em que as operações foram realizadas, que, segundo a regra apresentada anteriormente (PEMDAS), foi a seguinte:

```python
area = (B + b) * h / 2
print('Área do Trapézio: ', area)
```

Operadores de Comparação

```python
x = 5 
y = 1 
s = 'palavra' 

print('x > y:', x > y) 
print('x <= 4.564:', x <= 4.564) 
print('x <= (4.564 + 1):', x <= (4.564 + 1)) 
print('s == palavra:', s == 'palavra') 
print('s != palavra:', s != 'palavra') 
print('(y * 0) == False:', (y * 0) == False) 
print('s != abacaxi:', s != 'abacaxi')
```

```python
print(x == s) 
print(x != s) 
print(x > s)
```

Quando os operadores de maior (>), maior ou igual (>=), menor (<) e menor ou igual (<=) são aplicados em operandos do tipo string, a verificação ocorrerá pela comparação lexicográfica entre as duas strings. Ou seja, sequências que, na ordem alfabética, antecedem outras, são consideradas menores. Por exemplo:

```python
a = 'casa' 
b = 'predio' 
c = 'elevador' 
d = 'casamento' 
print(b < a) 
print(b > c)
print(a > c) 
print(a > d) 
print('abacaxi' >= 'abacaxi')
```
