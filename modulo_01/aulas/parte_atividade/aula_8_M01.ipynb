Conversão e Formatação dos Tipos de Dados

```python
nascimento = 1986 
ano_atual = 2022 
```

```python
idade = ano_atual - nascimento 
print('Eu tenho ' + idade + ' anos')
```

Isso ocorre porque a concatenação de strings com outros tipos de dados não é permitida.

Conversão dos tipos de Dados

```python
idade = ano_atual - nascimento 
print('Eu tenho ' + str(idade) + ' anos')
```

A seguir são apresentados alguns exemplos das funções de conversão de tipos:

```python
#conversão do tipo int 
print(float(10)) # 10.0 
print(bool(-1)) # True 
print(bool(0)) # False 

#conversão do tipo float 
print(int(9.999)) # 9 
print(bool(-0.99)) # True 
print(str(-0.99)) # '-0.99' 

#conversão do tipo bool 
print(int(True)) # 1 
print(float(False)) # 0.0
print(str(True)) # 'True' 

#conversão do tipo str 
print(int('-99')) # -99 
print(float('0.01')) # 0.01 
print(bool('palavra')) # True 
print(bool('')) # False
```

Operações de conversão com o tipo `NoneType`

```python
x = 10 
print('Valor:', x, '--', type(x)) 
# atribuição do valor None à variável x 
x = None 
print('Valor:', x, '--', type(x))
```

A conversão do valor None para outros tipos é possível apenas para os tipos `str` e `bool`.

```python
print(str(None)) 
print(bool(None))
```

```python
nascimento = 1986 
ano_atual = 2022 
idade = ano_atual - nascimento 
print(f'Eu tenho {idade} anos')
print(f'Eu tenho {ano_atual - nascimento} anos')
```

```python
orcamento = 3 
vlr_gasto = 10 
pct = (vlr_gasto/orcamento) * 100 
print(f'1. Porcentagem já gasta do orçamento: {pct}%')
print(f'2. Porcentagem já gasta do orçamento: {pct:.2f}%') # Formatação de casas decimais

pct = (vlr_gasto/orcamento)
print(f'3. Porcentagem já gasta do orçamento: {pct:.2%}') # Formatação de porcentagem com formatação de casas decimais
```
