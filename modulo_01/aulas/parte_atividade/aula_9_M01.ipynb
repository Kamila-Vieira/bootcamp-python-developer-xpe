Estruturas Condicionais

```python
# Exemplo em que a condição é verdadeira 
idade = 35 
if idade >= 18: 
    print('Idade suficiente para CNH!')

# Exemplo em que a condição é falsa 
idade = 15 
if idade >= 18: 
    print('Idade suficiente para CNH!')
```

```python
# Verifica se uma pessoa está apta para se aposentar 
idade = 70 
tempo_contrib = 20 
if idade >= 65 or tempo_contrib >= 35: 
    print('Habilitado para solicitar aposentadoria!')
```

```python
# Verifica se uma pessoa não está apta a se aposentar 
idade = 35 
tempo_contrib = 20 
if idade < 65 and tempo_contrib < 35: 
    print('Não habilitado para solicitar aposentadoria!')
```

Estrutura Condicional if – else

```python
# Exemplo em que a condição é verdadeira 
idade = 35 
if idade >= 18: 
    print('Idade suficiente para CNH!') 
else: 
    print('Idade não suficiente para CNH!')

# Exemplo em que a condição é falsa 
idade = 15 
if idade >= 18: 
    print('Idade suficiente para CNH!') 
else: 
    print('Idade não suficiente para CNH!')
```

```python
# Verifica se uma pessoa está apta para se aposentar 
idade = tempo_contrib = 20 
if idade >= 65 or tempo_contrib >= 35: 
    print('Habilitado para solicitar aposentadoria!') 
else: 
    print('Não habilitado para solicitar aposentadoria!')


if idade >= 65: 
    print('Habilitado para aposentadoria por idade!') 
else: 
    if tempo_contrib >= 35: 
        print('Habilitado para aposentadoria por tempo de contribuição!') 
    else: 
        print('Não habilitado para solicitar aposentadoria!')
```

Instrução Condicional if – elif – else

```python
# idade 22 anos, faixa etária: Adulto (if - else)
idade = 22 

if idade < 12: 
    faixa_etaria = 'Criança' 
else: 
    if idade < 18: 
        faixa_etaria = 'Adolescente' 
    else: 
        if idade < 60: 
            faixa_etaria = 'Adulto' 
        else: 
            faixa_etaria = 'Idoso' 

print('Faixa Etária 22: ', faixa_etaria, '(if - else)')

# idade 10 anos, faixa etária: Criança (if - elif - else)
idade = 10

if idade < 12: 
    faixa_etaria = 'Criança'
elif idade < 18:
    faixa_etaria = 'Adolescente' 
elif idade < 60:
    faixa_etaria = 'Adulto'
else:
    faixa_etaria = 'Idoso'

print('Faixa Etária 10: ', faixa_etaria, '(if - elif - else)')
```
