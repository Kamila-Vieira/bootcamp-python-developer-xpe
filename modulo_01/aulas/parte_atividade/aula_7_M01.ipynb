Operadores de concatenação (+) e repetição (*) de strings

```python
s1 = 'Belo' 
s2 = 'Horizonte' 
# Concatenação (+) 
print(s1 + s2) 
print(s1 + ' ' + s2) 
print(s1 + ' Monte') 
# Repetição print(s1 * 5) 
print((s1 + ', ') * 4)
```

Operadores de verificação de strings

```python
s1 = 'consolação' 
s2 = 'sola' 
print(s1 in s2) 
print(s2 in s1) 
print('solar' in s1) 
print('solar' not in s2)
```

Métodos das Strings

```python
palavra = 'consolação' 
print(len(palavra))
print(palavra.upper()) # todas as letras maiúsculas 
print(palavra.lower()) # todas as letras minúsculas 
print(palavra.title()) # primeira letra de cada palavra em maísculo

print(palavra.replace('consol', 'cor')) # substitui 'consol' por 'cor' 
print(palavra.startswith('consol')) # verifica se a string começa com 'consol' 
print(palavra.endswith('ação')) # verifica se a string termina com 'ação' 
print(palavra.find('n')) # retorna o índice da primeira ocorrência da letra n 
print(palavra.split()) # particiona a string em outras, que são separadas por um espaço 
print(palavra.split('l')) # particiona a string em outras, que são separadas por n 
palavra = ' palavra com espaços '
print(palavra.strip()) # remove os espaços extras no início e fim da string
```
