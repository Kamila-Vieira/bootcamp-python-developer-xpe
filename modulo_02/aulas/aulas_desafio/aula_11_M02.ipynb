# Scikit-learn

```python
import pandas as pd
import numpy as np
```

```python
# leitura dos dados csv

df = pd.read_csv("./temperature.csv")
df
```

## Classificação no Scikit-learn

```python
# extração de x e y

x, y = df[['temperatura']].values, df[['classification']].values

print('x:\n', x)
print('y:\n', y)
```

```python
# pré-processamento
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder?
```

```python
# conversão de y para valores numéricos
le = LabelEncoder() # label enconder
y = le.fit_transform(y.ravel())

print("y:\n", y)
```

```python
# modelo
from sklearn.linear_model import LogisticRegression
```

```python
# classificador => Apresenta os melhores resultados de amostras e atributos para determinada classe
clf = LogisticRegression()
clf.fit(x, y) # Localiza os melhores coeficientes da função para representar as classes (y)

# x => Conjunto de amostras e atributos
# x => Váriaveis alvo
```

```python
# score do modelo => Métrica que valida a qualidade do modelo para determinada base de dados

score = clf.score(x, y)
print("score:", score)
```

```python
# gerando 100 valores de temperatura
# linearmente espacados entre 6 e 45
# predicdo em novos valores de temperatura
x_test = np.linspace(start=0., stop=45., num=100).reshape(-1, 1)

# predicado desses valores
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
# conversão de y pred para os valores originais
y_pred_inv = le.inverse_transform(y_pred)
```

```python
y_pred_inv
```

```python
x_test
```

```python
# output => Dados de amostra num intervalo de valores gerados a partir do aprendizado de máquina
output = {'new_temp': x_test.ravel(), 
          'new_class': y_pred_inv.ravel()}
output = pd.DataFrame(output)

output.head()
```

```python
# estatísticas

output.info()
```

```python
# estatísticas

output.describe()
```

```python
# contagem de valores gerados

output['new_class'].value_counts().plot.bar(figsize=(10, 5), rot=0, title="# de novos valores gerados");
```

```python
# distribuição do output produzido
# conseguimos inferir a classificação de novas temperaturas
# a partir de um dataset com 6 exemplos

output.boxplot(by='new_class', figsize=(10, 5));
```

```python
# sistema automático de classificação de temperatura
def classify_temp():
    """Classifica o input do usudrio."""

    ask = True

    while ask:
        # input de temperatura
        temp = input("Insira a temperatura (graus Celsius):")

        # transformar para numpy array
        temp = np.array(float(temp)).reshape(-1, 1)

        # realiza classificacéo
        class_temp = clf.predict(temp)

        # transformacéo inversa para retornar a string original
        class_temp_inv = le.inverse_transform(class_temp)

        # classificacao
        print(f"A classificão da temperatura {temp.ravel()[0]} é:", class_temp_inv[0])

        # perguntar
        ask = input("Nova classificacéo (y/n): ") == 'y'
```

```python
classify_temp()
```
