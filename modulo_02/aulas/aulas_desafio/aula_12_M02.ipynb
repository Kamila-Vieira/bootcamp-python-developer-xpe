## Regressão linear no scikit-learn - I

```python
import numpy as np
import pandas as pd

# visualização de dados
import matplotlib.pyplot as plt

# pré-processamento de dados
from sklearn.preprocessing import LabelEncoder

# modelo de dados
from sklearn.linear_model import LogisticRegression
```

```python
le = LabelEncoder()
clf = LogisticRegression()
```

```python
# dados do segundo capitulo

x = [-1., -0.77777778, -0.55555556, -0.33333333, -0.11111111,
     0.11111111, 0.33333333, 0.55555556, 0.77777778, 1.]
     
y = [-1.3956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657,
     1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684202]
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
# transformando em numpy array
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)
```

```python
# modelo regressão linear
from sklearn.linear_model import LinearRegression
```

```python
# treinando o modelo: y = a*x + b, valores reais (a, b) = (2, 1)
reg = LinearRegression()
reg.fit(x, y) # Gera a melhor função para descrever os dados
```

```python
# coeficientes a, b estimados:
# valores estimados usando o numpy diretamente
# a estimado no numpy (aula 7): 2.1170001339465565
# b estimado no numpy (aula 7): 0.942383453

print("a estimado:", reg.coef_.ravel()[0])
print("b estimado:", reg.intercept_[0])
```

```python
# score do modelo => Métrica que valida a qualidade do modelo para determinada base de dados (0 à 1)

score = reg.score(x, y)
print("score:", score)
```

```python
# predição do modelo

y_pred = reg.predict(x)
```

```python
# plot dos dados

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, y_pred, label='regresão Linear (R2: {:.3f})'.format(score))
plt.legend()
plt.xlabel ("x")
plt.ylabel ("y")
plt.title("Regressdo Linear no scikit-learn")
plt.grid()
plt.show()
```
