## Criação de arrays

```python
# Criação de array 1D: [1, 2, 3]
l = [1, 2, 3]
x = np.array(l)
print('x:', x)
print('shape:', x.shape) 
```

```python
type(x)
```

```python
# Criação de array 2D: [[1, 2], [3, 4]] (Listas aninhadas)
dim = [[1, 2], [3, 4]]
y = np.array(dim)
print('y:', y)
print('shape:', y.shape) 
```

Operações com numpy

```python
# Array contendo apenas zeros (0's)
# zeros(shape, dtype=float, order='C', *, like=None)

sp = (3, 3) # (linhas, colunas)
z = np.zeros(sp, dtype=int)
print('y:', z)
print('shape:', z.shape) 
```

```python
help(np.zeros)
```

```python
# Array contendo apenas ums (1's)
# ones(shape, dtype=None, order='C', *, like=None)

sp = (2, 3) # (linhas, colunas)
z = np.ones(sp, dtype=int)
print('y:', z)
print('shape:', z.shape) 
```

```python
help(np.ones)
```

```python
# crray com valores dentro de um intervalo com distância entre eles igualmente espaçadas
# linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

# valores uniformemente espaçados entre 5 e 15
x_min, x_max = 5, 15
ls = np.linspace(start=x_min, stop=x_max, num=6)
print('ls:', ls)
print('shape:', ls.shape) 
```

```python
help(np.linspace)
```

```python
# Criação de matriz identidade: Array diagonal composta apenas por 1's e demais elementos compostos por 0's
# eye(N, M=None, k=0, dtype=<class 'float'>, order='C', *, like=None)

n = 4
mi = np.eye(n)
print('mi:', mi)
print('shape:', mi.shape) 
```

```python
help(np.eye)
```

```python
# Array de valores aleatórios entre 0 e 1
# random(size=None)

rd = np.random.random(size=(3, 2))
print('rd:\n', rd)
print('shape rd:', rd.shape) 

print() 

rd2 = np.random.random(5)
print('rd2:', rd2)
print('shape rd2:', rd2.shape) 
```

```python
help(np.random.random)
```
