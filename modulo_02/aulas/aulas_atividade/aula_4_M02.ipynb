````python
# importando as bibliotecas
import numpy as np
````

## Operações aritméticas

Operações elemento a elemento (operações entre arrays de tamanhos equivalentes)

````python
# Criação de dois arrays x e y

x = np.ones((2, 2), dtype=int) # matriz 2x2 de 1's
y = np.eye(2, dtype=int) # matriz identidade

print('x:\n', x)
print('\ny:\n', y)
````

````python
# soma

print('Soma x com y:\n', x + y)
````

````python
# subtração

print('subtração x - y:\n', x - y)
print('\nsubtração y - x:\n', y - x)
````

````python
# divisão

print('subtração y / x:\n', y / x)
print('\nsubtração x / y:\n', x / y) # Retorna um erro de divisão por zero
````

````python
# multiplicação

print('multiplicação x com y:\n', x * y)
````

Operações Broadcasting (operações entre arrays de tamanhos diferentes)

````python
# soma

print('Soma x com 2:\n', x + 2) # equivale à [[1 1] [1 1]] + [[2 2] [2 2]]
print('\nSoma y com 2:\n', y + 2) # equivale à [[1 0] [0 1]] + [[2 2] [2 2]]
````

````python
# subtração

print('subtração x - 2:\n', x - 2) # equivale à [[1 1] [1 1]] - [[2 2] [2 2]]
print('\nsubtração y - 2:\n', y - 2) # equivale à [[1 0] [0 1]] - [[2 2] [2 2]]
print('\nsubtração 2 - x:\n', 2 - x) # equivale à [[2 2] [2 2]] - [[1 1] [1 1]]
print('\nsubtração 2 - y:\n', 2 - y) # equivale à [[2 2] [2 2]] - [[1 0] [0 1]]
````

````python
# divisão

print('subtração x / 2:\n', x / 2) # equivale à [[1 1] [1 1]] / [[2 2] [2 2]]
print('\nsubtração y / 2:\n', y / 2) # equivale à [[1 0] [0 1]] / [[2 2] [2 2]]
print('\nsubtração 2 / x:\n', 2 / x) # equivale à [[2 2] [2 2]] / [[1 1] [1 1]]
print('\nsubtração 2 / y:\n', 2 / y) # equivale à [[2 2] [2 2]] / [[1 0] [0 1]] # Retorna um erro de divisão por zero
````

````python
# multiplicação

print('multiplicação x com 2:\n', x * 2) # equivale à [[1 1] [1 1]] * [[2 2] [2 2]]
print('\nmultiplicação y com 2:\n', y * 2) # equivale à [[1 0] [0 1]] * [[2 2] [2 2]]
````

````python
# array 10x10 contendo apenas 10

print(10 * np.ones((10, 10), dtype=int))

# OU

print('\n', 10 + np.zeros((10, 10), dtype=int))
````

````python
# quando o broadcast não funciona

np.array([1, 2, 3]) + np.array([1, 2]) # Gera uma inconsistência entre os array 
````

````python
# Soma, subtração e divisão
print('Combinação entre operações:\n', (x + y) / (x - 2))

# [[1 1] [1 1]] + [[1 0] [0 1]] = [[2 1] [1 2]] => 
# [[1 1] [1 1]] - [[2 2] [2 2]] = [[-1 -1] [-1 -1]] => 
# [[2 1] [1 2]] / [[-1 -1] [-1 -1]] => [[-2. -1.] [-1. -2.]]
````

Operações matriciais (multiplicação de matrizes)

````python
# multiplicação matricial

print('multiplicação matricial (np.dot):\n', np.dot(x, y) )
print('\nmultiplicação matricial (@):\n', x @ y)
print('\nmultiplicação matricial (.dot):\n', x.dot(y))

# multiplicação matricial com broadcasting

print('\nmultiplicação matricial com broadcasting (np.dot):\n', np.dot(x, 2) )
print('\nmultiplicação matricial com broadcasting (.dot):\n', x.dot(2))
````

```
    Exemplo:
    Solução de um sistema de equações:
    a + 2b = 7 ===> 1*a + 2*b = 7
    3a - 2b = -11 ===> 3*a - 2*b = -11
    solucão analítica: (a, b) = (-1, 4)

    Matricialmente, este problema tem a seguinte forma:
    Ax = c, onde:
    - x = [[a], [b]]
    - A = [[1, 2], [3, -2]]
    - c = [[7], [-11]]
    solução numérica: x = inv(A) @ c.
```

````python
# definição do problema
A = np.array([[1, 2], [3, -2]])
c = np.array([[7], [-11]])
print('A:\n', A)
print('c:\n', c)
````

````python
help(np.linalg.inv)
````

````python
# Solução
x = np.dot(np.linalg.inv(A), c) # x = inv(A) @ c
print('(a, b):', x.ravel())
````
