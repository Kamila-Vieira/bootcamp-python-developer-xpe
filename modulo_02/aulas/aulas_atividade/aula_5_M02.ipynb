```python
import numpy as np
```

## Comparações (operações elemento a elemento)

```python
# criação dos arrays 
x = np.array([[1,2], [3,4]])
y = np.array([1.5, 3.5])

print('x:\n', x)
print('y:\n', y)
```

```python
# comparações ponto a ponto
print('Comparação de um array com escalar (>):\n', x > 2)
print('\nComparação de um array com escalar (>=):\n', x >= 2)
print('\nComparação de um array com escalar (<):\n', x < 2)
print('\nComparação de um array com escalar (<=):\n', x <= 2)

# comparações broadcasting
print('\nComparação entre arrays (==):\n', x == y)
print('\nComparação entre arrays (>):\n', x > y)
print('\nComparação entre arrays (<):\n', x < y)
```

## Indexação booleana (operações de filtro)

```python
# criação de mu array

x = np.array([[1, 3, 7],
              [4, 11, 21],
              [42, 8, 9]])
print('x:\n', x)
```

```python
# Retornar o número de elementos maiores que k

k = 10
cond = x > k
print('condição:\n', cond)
print(f'\nelementos maiores que {k}:', x[cond])
print(f'\nnúmero de elementos maiores que {k}:', len(x[cond]))
```

```python
# extração de números pares

cond = x % 2 == 0
print('condição:\n', cond)
print('\nnúmeros pares:', x[cond])
```

```python
# extração de números ímpares

cond = x % 2 == 1
print('condição:\n', cond)
print('\nnúmeros ímpares:', x[cond])
```
