## Regressão linear no Numpy

Encontrar uma função que melhor se aproxima do intervalo de tempo (menor erro quadrático) entre os dados

```python
# dados

x = [-1., -0.77777778, -0.55555556, -0.33333333, -0.11111111,
     0.11111111, 0.33333333, 0.55555556, 0.77777778, 1.]
y = [-1.3956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657,
     1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684202]
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
"""
Iremos estimar uma função do tipo: y = a*x + b
ou seja, devemos achar quais os valores de aeb
que melhor representam os dados.

Os valores reais de a e b são: (a, b): 2, 1
"""

# transformando para numpy e vetor coluna
х = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)

# adicionando bias: para estimar 0 termo b
X = np.hstack((х, np.ones(х.shape)))

# estimando a e b
beta = np.linalg.pinv(X).dot(y) # OU np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y) == inv(X.T * X) * X.T * y (a inversa da transporta de X vezes X (inv(X.T * X)), vezes a transporta de X (X.T), vezes y)
print("a estimado:", beta[0][0])
print("b estimado:", beta[1][0])
```

```python
# plot dos dados
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='dados originais')
plt.plot(x, X.dot(beta), label='regressão linear')
plt.legend()
plt.xlabel ("x")
plt.ylabel("y")
plt.title("Regressão Linear com numpy")
plt.grid()
plt.show()
```
