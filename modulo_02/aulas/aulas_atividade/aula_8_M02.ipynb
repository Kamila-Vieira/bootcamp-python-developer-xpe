```python
# leitura de planilhas excel
# 2 abas (worksheets)

# leitura do arquivo todo
excel_file = pd.ExcelFile("./temperature.xlsx")
```

```python
type(excel_file)
```

```python
# leitura da primeira aba (Sheet1)
# dados numéricos com separador decimal='.'

df1 = pd.read_excel(excel_file, sheet_name='Sheet1')
df1
```

```python
pd.read_excel?
```

```python
def decimal_converter(value):
    try:
        return float(value.replace(',', '.'))
    except ValueError:
        return value
```

```python
# leitura da segunda aba (Sheet1)
# dados numéricos com separador decimal=','
df2 = pd.read_excel(excel_file, sheet_name='Sheet2', converters={ 'temperatura': decimal_converter }) # decimal depreciado
df2
```
