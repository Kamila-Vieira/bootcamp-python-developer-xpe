axis => representam os eixos dos arrays:
 
- Array 1D => axis 0 (linhas)
- Array 2D => axis 0 (linhas), axis 1 (colunas)
- Array 3D => axis 0 (linhas), axis 2 (colunas), axis 3 (profundidade)

shape => Quantidades de elementos em cada eixo (axis):

- Array 1D => shape(x,) (elementos linha,)
- Array 2D => shape(x, y) (elementos linha, elementos coluna)
- Array 3D => shape(x, y, z) (elementos linha, elementos coluna, elementos profundidade)
