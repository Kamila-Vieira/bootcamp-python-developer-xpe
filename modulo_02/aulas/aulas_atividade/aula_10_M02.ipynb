```python
import pandas as pd
```

```python
# leitura dos dados csv

df = pd.read_csv("./temperature.csv")
df
```

```python
# transformando o tipo da coluna date para datetime

df['date'] = pd.to_datetime(df['date'])
```

```python
# setando o indice

df = df.set_index('date')
```

```python
# contagem de dados

df['classification'].value_counts()
```

## Ordenação no Pandas

### Método sort_values()

```python
# ordenação crescente por uma coluna (ordem crescente de números)

df.sort_values(by="temperatura")
```

```python
# ordenação crescente por uma coluna (ordem crescente alfabética)

df.sort_values(by='classification')
```

```python
# ordenação crescente por uma coluna (ordem crescente primeira coluna primeiro)

df.sort_values(by=['classification', 'temperatura'])
```

```python
# ordenação decrescente por uma coluna (ordem decrescente de números)

df.sort_values(by="temperatura", ascending=False)
```

```python
# ordenação decrescente por uma coluna (ordem decrescente primeira coluna primeiro)

df.sort_values(by=['classification', 'temperatura'], ascending=False)
```

### Método sort_index()

```python
# ordenação crescente por indice

df.sort_index()
```

```python
# ordenação decrescente por indice

df.sort_index(ascending=False)
```

## Visualização de dados no Pandas

```python
# plot de linhas

df.plot();
```

```python
# plot de linhas: tamanho

df.plot(figsize=(10, 5));
```

```python
# plot de linhas: grid

df.plot(figsize=(10, 5), grid=True);
```

```python
# plot de linhas: style

df.plot(figsize=(10, 5), grid=True, style="-o"); # pontos com linha
# df.plot(figsize=(10, 5), grid=True, style="o"); # pontos
# df.plot(figsize=(10, 5), grid=True, style="--"); # tracejado
# df.plot(figsize=(10, 5), grid=True, style="-."); # tracejado com pontinhos
```

```python
# plot de linhas: linewidth

df.plot(figsize=(10, 5), grid=True, style="-o", linewidth=2.5);
```

```python
# plot de linhas: color

# df.plot(figsize=(10, 5), grid=True, style="-o", linewidth=2.5, color='red');
df.plot(figsize=(10, 5), grid=True, style="-o", linewidth=2.5, color='#008888'); # com hexadecimal
```

```python
# plot de barras

df.plot(figsize=(10, 5), kind="bar", rot=30);
```

```python
# plot de barras com contagem de valores

df['classification'].value_counts().plot(figsize=(10, 5), kind="bar", rot=0);
```

```python
# plot de pizza (pie plot)

df['classification'].value_counts().plot.pie(autopct="%1.1f%%", figsize=(10, 5), shadow=True);
```
