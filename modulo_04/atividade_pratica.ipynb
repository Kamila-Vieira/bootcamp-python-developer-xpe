2. Para a separação dos dados de treino e teste considere usar o módulo ``train_test_split``

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard_scaler = StandardScaler()

x = standard_scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rede_neural = MLPClassifier(hidden_layer_sizes=(5,5), max_iter=500, alpha=0.0001, learning_rate_init=0.001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.000000001, momentum=0.01)
```
