```python
import numpy as np
```

```python
x = np.array([[160], [165], [171], [174], [179], [181], [185], [188], [191], [200]]) # altura dos funcionrios
```

```python
y = np.array([[64], [67], [70], [80], [77], [81], [87], [94], [101], [112]]) # peso dos funcionrios
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
# prever quantos quilos pode ter um funcionrio que possui 1.90m

previsao = regressor.intercept_ + regressor.coef_ * 190 

previsao
```

```python
# Previso acima pode ser feita tambm da seguinte forma

altura = np.array([190]).reshape(1, 1)

previsao2 = regressor.predict(altura)

previsao2
```

```python
# Previso de peso de todos os funcionrios

previsoes = regressor.predict(x)

previsoes
```

```python
# Diferena/margem de erro absoluta da previso

resultado = abs(y - previsoes)

resultado
```

```python
# Mdia da diferena/margem de erro da previso

resultado.mean()
```

```python
#  possvel tambm fazer as mtricas de margem de erro da seguinte forma

from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# mdia da margem de erro absoluta

mae = mean_absolute_error(y, previsoes)

mae
```

```python
# mdia da margem de erro ao quadrada

mse = mean_squared_error(y, previsoes)

mse
```

```python
plt.plot(x, y, 'o');
plt.plot(x, previsoes, color="red");
plt.title("Regresso Linear Simples");
plt.xlabel('Altura');
plt.ylabel('Peso');
```
