```python
import numpy as np
```

```python
x = np.array([[160], [165], [171], [174], [179], [181], [185], [188], [191], [200]]) # altura dos funcionários
```

```python
y = np.array([[64], [67], [70], [80], [77], [81], [87], [94], [101], [112]]) # peso dos funcionários
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
# prever quantos quilos pode ter um funcionário que possui 1.90m

previsao = regressor.intercept_ + regressor.coef_ * 190 

previsao
```

```python
# Previsão acima pode ser feita também da seguinte forma

altura = np.array([190]).reshape(1, 1)

previsao2 = regressor.predict(altura)

previsao2
```

```python
# Previsão de peso de todos os funcionários

previsoes = regressor.predict(x)

previsoes
```

```python
# Diferença/margem de erro absoluta da previsão

resultado = abs(y - previsoes)

resultado
```

```python
# Média da diferença/margem de erro da previsão

resultado.mean()
```

```python
# É possível também fazer as métricas de margem de erro da seguinte forma

from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# média da margem de erro absoluta

mae = mean_absolute_error(y, previsoes)

mae
```

```python
# média da margem de erro ao quadrada

mse = mean_squared_error(y, previsoes)

mse
```

```python
plt.plot(x, y, 'o');
plt.plot(x, previsoes, color="red");
plt.title("Regressão Linear Simples");
plt.xlabel('Altura');
plt.ylabel('Peso');
```
